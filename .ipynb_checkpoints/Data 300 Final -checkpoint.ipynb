```python
# since the data is imbalanced, we must undersample it in order to 
# run more accurate models

not_bankrupt_b = resample(not_bankrupt, random_state=7, n_samples = 220, replace=True)
under_sampled_data = pd.concat([not_bankrupt_b, bankrupt])
under_sampled_data
```

## Part 1

```python
# this splits the data into training and test sets

y = original_data["Bankrupt"]
x = original_data.loc[:, original_data.columns != "Bankrupt"] 
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=7)
```

```python
# this creates a default Random Forest Classifier model, fits the model
# uses the model to predict which class the data is in, and tracks how
# long it took the model to run
```

```python
%%time

RF_model1 = RandomForestClassifier()
RF_model1.fit(X_train, y_train)
y_pred1 = RF_model1.predict(X_test)
y_pred1
```

```python
# this runs a confusion matrix and classification report so we can see 
# different error metrics of the model

print(confusion_matrix(y_test, y_pred1))

print(classification_report(y_test, y_pred1))
```

```python
# this creates a more specific Random Forest Classifier model with
# criterion = entropy, max_features=log2, and random_state=5, fits the
# model uses the model to predict which class the data is in, and tracks
# how long it took the model to run
```

```python
%%time

RF_model2 = RandomForestClassifier(criterion = 'entropy', max_features = 'log2', random_state = 5, )
RF_model2.fit(X_train, y_train)
y_pred2 = RF_model2.predict(X_test)
y_pred2
```

```python
# this runs a confusion matrix and classification report so we can see 
# different error metrics of the model

print(confusion_matrix(y_test, y_pred2))
print(classification_report(y_test, y_pred2))
```

2. I would prefer the default model, even though their error metrics were very similar, the default model did slightly better. Although the defualt model takes slightly longer, I think it is worthwhile to have a better model despite the time it takes to fit that model.

3. The default model takes longer to fit, only by a small amount though. The default model took 2.04 seconds and the non default model took 1.04 seconds

## Part 2

```python
# this splits the data into training and test sets

y = under_sampled_data["Bankrupt"]
x = under_sampled_data.loc[:, under_sampled_data.columns != "Bankrupt"] 
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=7)
```

```python
# this creates a default Random Forest Classifier model, fits the model
# uses the model to predict which class the data is in, and tracks how
# long it took the model to run
```

```python
%%time

RF_model = RandomForestClassifier()
RF_model.fit(X_train, y_train)
y_pred_RF = RF_model.predict(X_test)
y_pred_RF
```

```python
# this creates a default SVM Classifier model, fits the model
# uses the model to predict which class the data is in, and tracks how
# long it took the model to run
```

```python
%%time

SVM_model = SVC()
SVM_model.fit(X_train, y_train)
y_pred_SVM = SVM_model.predict(X_test)
y_pred_SVM
```

1. Both of the fitting times are very fast, but the SVM model beat the Random Forest. The SVM fitted the model in 18.5 ms and the Random Forest fit the model in 130 ms.

```python
# this runs a confusion matrix and classification report so we can see 
# the different error metrics of the models for comparison


print(confusion_matrix(y_test, y_pred_RF))
print(confusion_matrix(y_test, y_pred_SVM))

print(classification_report(y_test, y_pred_RF))
print(classification_report(y_test, y_pred_SVM))
```

2. The Random Forest Model does a much better job of predicting the classes correctly. This is seen in all of the error metrics I used, however it does take slightly longer. For me even though it takes longer, it is only miliseconds so I would still prefer the Random Forest Model.
