```python
# this function runs the KPSS test on both columns and returns the KPSS
#test statistic, the p-value of the test, the number of lags used, and 
#the critical values 

for i in ep.columns:
    print (i)
    print()
    print (stattools.kpss(ep[i], regression='ct'))
    print ()
```

```python
#this function runs the ADF test on both columns in the data set and 
#returns the ADF test statistic, the p-value of the test, the number of
#lags used, the number of observations used to complete the test, and
#the critical values

for i in ep.columns:
    print(i)
    print()
    print(adfuller(ep[i]))
    print()
```

2. The test statistic for the KPSS test for the date is 0.01715323608505754 with a p-value of 0.1, and the ADF test statistic for the date is -0.14017905755825988 with a p-value of 0.9452865588327521. The test statistic for the KPSS test for IPG2211A2N is 0.5051030832418347 with a p-value of 0.01, and the ADF test statistic for IPG2211A2N is -2.2569903500472432 with a p-value of 0.18621469116586814. From this we can conclude that there is a very good chance that the two variables are not stationary; the ADF test and the KPSS tests have very different values, both of which lean towards showing non-stationarity, but we can also look at the plots before deciding further.

```python
#this function removes the dates column as it is not needed for the 
#differncing section

ep_no_dates=ep[['IPG2211A2N']]
ep_no_dates
```

```python
#this function plots the raw data so we can see if it is stationary
#or not, to me there is a clear trend so i'd say it is not stationary

for i, col in enumerate(ep_no_dates.columns):
    ep_no_dates[col].plot(fig=plt.figure(i))
    plt.title(col)
    
plt.show()
```

```python
#this function differences the data once, and brings it to a more 
#stationary state

for i, col in enumerate(ep_no_dates.columns):
    ep_no_dates[col].diff().plot(fig=plt.figure(i))
    plt.title(col)

    plt.show() 
```

```python
#this function differences the data a third time, and bring the data 
#even closer to stationarity

for i, col in enumerate(ep_no_dates.columns):
    ep_no_dates[col].diff().diff().plot(fig=plt.figure(i))
    plt.title(col)

plt.show() 
```

3. The first, undifferenced graph has a very clear pattern, it seems to curve upward and to the right, while the second two have much less of a pattern. By the third graph there not only seems to be less of a pattern but the data also seems more centralized. 

```python
# this function creates ACF and PACF plots that are able to be shown 
#one after the other

def acf_pacf (series)-> plt.plot:
    return plot_acf(series), plot_pacf(series)
```

```python
#this runs through both variables in the data and creates ACF and PACF
#plots for each of them

for i in ep.columns:
    print(i)
    acf_pacf(ep[i])
    print()
```

```python
# this function creates and finds the optimal ARIMA model 
#for the time series data given

def arimamodel(timeseries):
    automodel = pm.auto_arima(timeseries, 
                              start_p=1, 
                              start_q=1,
                              test="adf",
                              seasonal=True,
                              trace=True)
    return automodel
```

#### For this class we said we were not supposed to use Auto ARIMA. And even if you used it, you use it on a stationary series. (-2)

```python
#this finds the optimal ARIMA model for the IPG2211A2N data, which is
#ARIMA(4,0,2)

for i in ep_no_dates:
    print(i)
    arimamodel(ep_no_dates[i])
    print()
```

5. My final ARIMA order is ARIMA(4,0,2)

#### How did you decide your ARIMA is of order 4,0,2? (-2). If the data is non stationary, it needs to be differenced at least once. 

```python
# This function models the data off of your chosen test set, which I set
# at 80% of the data, and finds how well the model fits the data

model = sm.tsa.arima.ARIMA(ep['IPG2211A2N'].head(int(0.8*len(ep))), order=(4, 0, 2))
results = model.fit()
```

```python
# this function shows us the summary of our ARIMA model and 
#its coefficients

results.summary()
```

```python
#This function forecasts out using the chosen 80% test set to create
# a predicted model

ep['IPG2211A2N_forecast']=results.predict(start = int(0.8*len(ep))+1, end=len(ep) , dynamic=True)
```

```python
#this function plots the actual data along with the forecasted model

ep[['IPG2211A2N','IPG2211A2N_forecast']].plot(figsize=(12,8))
```

```python
#these functions find the mean absolute error and the mean squared error
# which we can use to decide if the model is a good fit or not

MAE = mean_absolute_error(ep['IPG2211A2N_forecast'].tail(int(0.2*len(ep))),ep['IPG2211A2N'].tail(int(0.2*len(ep))))
MSE = mean_squared_error(ep['IPG2211A2N_forecast'].tail(int(0.2*len(ep))),ep['IPG2211A2N'].tail(int(0.2*len(ep))))
print("The mean absolute error is " + str(MAE))
print("The mean squared error is " + str(MSE))
```
