```python
y = original_data["Bankrupt"]
x = original_data.loc[:, original_data.columns != "Bankrupt"] 
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=7)
```

```python
RF_model1 = RandomForestClassifier()
RF_model1.fit(X_train, y_train)
y_pred1 = RF_model1.predict(X_test)
y_pred1
```

```python
RF_model2 = RandomForestClassifier(criterion = 'entropy', max_features = 'log2', random_state = 5, )
RF_model2.fit(X_train, y_train)
y_pred2 = RF_model2.predict(X_test)
y_pred2
```
